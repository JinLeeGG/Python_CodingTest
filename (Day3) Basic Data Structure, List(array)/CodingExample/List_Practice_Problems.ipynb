```python
# My answer
arr = [1, -5, 2, 4, 3]

def solution(arr):
  return sorted(arr)

solution(arr)
```

## 문제] 정수 배열을 하나 받는다. 그 중 배열의 중복값을 제거하고 배열 데이터를 내림차순으로 정렬해서 반환하는 함수 작성하시오

입력
[4, 2, 2, 1, 3, 4]

결과값으로는
[4, 3, 2, 1]

```python
# My answer
import numpy as np

arr = [4, 2, 2, 1, 3, 4]

def solution(arr):
  return list(set(arr))
  # reverse 하게 해주는 함수 추가

solution(arr)
```

```python
import numpy as np

arr = [4, 2, 2, 1, 3, 4]

def solution(arr):
  list1 = list(set(arr))
  list1.sort(reverse = True)
  return list1

solution(arr)
```

## 문제)
정수 배열 num이 주어질때 리스트 num에서 서로 다른 인덱스에 있는 2개의 수를 뽑아 더해 만들 수 있는  모든 수를 배열에 오름차순으로 정렬하는 solution을 작성하시오.

예를들어 [2, 1, 3, 4, 1] ==>

* 합의 수가 같으면 중복 제거 한다.


```python
# My answer
def solution(num):
    ret = set()  # 중복을 제거하기 위해 set 사용
    for i in range(len(num)):
        for j in range(i + 1, len(num)):  # i+1부터 시작하여 같은 인덱스 사용 방지
            ret.add(num[i] + num[j])     # 두 수의 합을 set에 추가
    return sorted(ret)                   # 오름차순 정렬된 리스트로 반환

# 테스트
arr = [2, 1, 3, 4, 1]
result = solution(arr)
print(result)  # 출력: [2, 3, 4, 5, 6, 7]
```

```python
answers = [1,2,3,4,5]
answers1 = [1,3,2,4,2]

def solution(answers):
    man1 = [1,2,3,4,5]                   # 패턴을 나타내는 리스트
    man2 = [2,1,2,3,2,4,2,5]
    man3 = [3,3,1,1,2,2,4,4,5,5]
    score = [0, 0 ,0]
    result = []

    for idx, answer in enumerate(answers):      #리스트 안에 들어있는 index와 값을 둘다 리턴해준다.
        if answer == man1[idx%len(man1)]:
            score[0] += 1
        if answer == man2[idx%len(man2)]:
            score[1] += 1
        if answer == man3[idx%len(man3)]:
            score[2] += 1

    for idx, s in enumerate(score):
        if s == max(score):
            result.append(idx + 1)

    return result

solution(answers1)
```
