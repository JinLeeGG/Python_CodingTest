```python
s1 = "(())()"
s2 = "((())()"

def solution(s):
    stack = []            # Create a new stack
    for i in s:           # go through each string
      if i == "(":        # If i == (
        stack.append(i)   # add to stack
      elif i == ")":      # If i == )
        if not stack:     # and If the stack is empty,
          return False    # return False because you cannot close ) with anything
        else:             # otherwise,
          stack.pop()     # it will close it '()' so pop the '('

    # if stack:
    #   return False
    # else:
    #   return True

    return len(stack) == 0   # If stack is empty, it sucessfully closed () which is True If not, False

solution(s1)
```

```python
s1 = 'baabaa'
s2 = 'cdcd'

def solution(s):
    stack = []                                # Create a new Stack
    for i in s:                               # go through each characters in string
        if stack and stack[-1] == i:          # if stack is not empty and last item in stack (stack[-1]) is equal to current character,
            stack.pop()                       # pop out the character in the stack becuse we want to remove any duplicates when each charactor is added in the stack
        else:                                 # Otherwise,
          stack.append(i)                     # If stack is empty or not empty, add current character in the stack

    if len(stack) == 0:                       # If the length of stack is 0, which we have succesfully removed all characters because they all have duplicates, return 1
      return 1
    else:
      return 0                                # Otherwise, there is at least one charactor in the stack that we were unable to remove, return 0.

solution(s2)
```

```python
def solution(prices):
    answer = [0] * (len(prices))  # Create a new list based on the length of prices

    for i in range(len(prices)):  # Iterate through each price in the list
        cnt = 0  # Initialize counter to track duration
        for j in range(i+1, len(prices)):  # Iterate over subsequent prices
            if prices[i] <= prices[j]:  # If current price is less than or equal to future price
                cnt += 1  # Increase counter
            else:
                cnt += 1  # Increase counter for the last comparison
                break  # Stop counting when price drops
        answer[i] = cnt  # Store the duration in the answer list

    return answer  # Return the final list of durations

# Test case
prices = [1, 2, 3, 2, 3]
print(solution(prices))  # Expected output: [4, 3, 1, 1, 0]
```
