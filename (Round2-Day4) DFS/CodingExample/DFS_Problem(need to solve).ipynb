```python

def solution(S, N, M):
    count = 0
    visited = [[False for _ in range(M)] for _ in range(N)]
    print(visited)

    for i in range(N):
        for j in range(M):
            if S[i][j] == '0' and visited[i][j] == False:
              count +=1

              explore(i, j, S, visited, N, M)

    return count

def explore(r, c, S, visited, N, M):
    # 얼음 틀 범위 벗어나는지 확인
    if r < 0 or r >=N or c <0 or c>= M:
        return

    # 현재 위치에 방문했거나 칸막이라면
    if visited[r][c] or S[r][c] == '1':
        return

    # 조건을 통과했을때는 아직 방문을 하지 않은 0이다.
    visited[r][c] = True

    # 상, 하, 좌, 우 확인
    explore(r - 1, c, S, visited, N, M) # 위쪽
    explore(r + 1, c, S, visited, N, M) # 아래쪽
    explore(r, c - 1, S, visited, N, M) # 왼쪽
    explore(r, c + 1, S, visited, N, M) # 오른쪽

S = [
 "00110",
 "00011",
 "11111",
 "00000"
]


solution(S, 4, 5)
```
