```python
# 사용자가 순차적으로 숫자들을 하나씩 선택

result = []

def backtrack(sum, selectNum, start):
    sum == 10
    result.append(selectNum)
    return

def solution(N):
  start = 1
  for i in range(start, N+1):
      if sum + i <= 10:  # 선택한 숫자의 합이 10보다 작거나 같으면
        backtrack(sum + i, selectNum + [i], i+1)

  backtrack(0, [], 1)
  return result

print(solution(10))
```
