```python
from collections import deque  # deque(덱) 자료구조를 사용하기 위해 import

def solution(board, r, c):
    answer = 0
    N = len(board)  # 보드의 크기
    directions = [(-1, 0), (0, -1), (1, 0), (0, 1)]  # 상, 좌, 하, 우 방향을 나타내는 튜플 리스트

    def bfs(x, y, cost, d):
        graph = [[0]*N for _ in range(N)]  # 각 칸까지의 최소 비용을 저장할 2차원 리스트 생성
        for a in range(N):
            for b in range(N):
                if board[a][b] > 0:  # 그림이 있는 칸이면
                    graph[a][b] = -1  # -1로 표시

        q = deque()  # BFS 탐색을 위한 큐 생성
        q.append((x, y, cost, d))  # 시작 위치(x, y), 현재까지 비용(cost), 방향(d)을 큐에 추가

        while q:  # 큐가 빌 때까지 반복
            x, y, cost, idx = q.popleft()  # 큐에서 현재 위치, 비용, 방향을 꺼냄
            for i in range(len(directions)):  # 4방향(상, 좌, 하, 우) 모두 탐색
                nx = x + directions[i][0]  # 다음 x좌표 계산
                ny = y + directions[i][1]  # 다음 y좌표 계산

                if nx < 0 or nx >= N or ny < 0 or ny >= N:  # 보드 범위를 벗어나면
                    continue  # 무시하고 다음 방향 탐색

                if graph[nx][ny] == -1:  # 그림을 발견하면
                    newcost = cost + 1

                if idx == i:  # 이전 방향과 같은 방향으로 이동하면(직선)
                    newcost = cost + 1


                # 아직 방문하지 않았거나, 더 적은 비용으로 도달할 수 있으면
                if graph[nx][ny] == 0 or ((graph[nx][ny] != 0) and graph[nx][ny] > newcost):
                    q.append((nx, ny, newcost, i))  # 다음 위치, 비용, 방향을 큐에 추가
                    graph[nx][ny] = newcost  # 최소 비용 갱신
                else:
                    continue  # 더 좋은 경로가 아니면 무시

        # 도착지가 여러번 옮겨질때?
        #????

        return graph[N-1][N-1]  # 도착지점까지의 최소 비용 반환

    # r, c 를 기준으로 상 하 좌 우 방향으로 갔을때 최솟값 찾기
    return min(bfs(r, c, 0, 2), bfs(r, c, 0, 3), bfs(r, c, 0, 4), bfs(r, c, 0, 0))

# 예시 입력 실행
solution([[1,0,0,3], [2,0,0,0], [0,0,0,2], [3,0,1,0]], 1, 0)
```
