```python
from collections import deque  # deque()    import

def solution(board, r, c):
    answer = 0
    N = len(board)  #  
    directions = [(-1, 0), (0, -1), (1, 0), (0, 1)]  # , , ,     

    def bfs(x, y, cost, d):
        graph = [[0]*N for _ in range(N)]  #      2  
        for a in range(N):
            for b in range(N):
                if board[a][b] > 0:  #   
                    graph[a][b] = -1  # -1 

        q = deque()  # BFS    
        q.append((x, y, cost, d))  #  (x, y),  (cost), (d)  

        while q:  #    
            x, y, cost, idx = q.popleft()  #   , ,  
            for i in range(len(directions)):  # 4(, , , )  
                nx = x + directions[i][0]  #  x 
                ny = y + directions[i][1]  #  y 

                if nx < 0 or nx >= N or ny < 0 or ny >= N:  #   
                    continue  #    

                if graph[nx][ny] == -1:  #  
                    newcost = cost + 1

                if idx == i:  #     ()
                    newcost = cost + 1


                #   ,      
                if graph[nx][ny] == 0 or ((graph[nx][ny] != 0) and graph[nx][ny] > newcost):
                    q.append((nx, ny, newcost, i))  #  , ,   
                    graph[nx][ny] = newcost  #   
                else:
                    continue  #     

        #   ?
        #????

        return graph[N-1][N-1]  #    

    # r, c          
    return min(bfs(r, c, 0, 2), bfs(r, c, 0, 3), bfs(r, c, 0, 4), bfs(r, c, 0, 0))

#   
solution([[1,0,0,3], [2,0,0,0], [0,0,0,2], [3,0,1,0]], 1, 0)
```
