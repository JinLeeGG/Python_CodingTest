```python

def solution(S, N, M):
    count = 0
    visited = [[False for _ in range(M)] for _ in range(N)]
    print(visited)

    for i in range(N):
        for j in range(M):
            if S[i][j] == '0' and visited[i][j] == False:
              count +=1

              explore(i, j, S, visited, N, M)

    return count

def explore(r, c, S, visited, N, M):
    #     
    if r < 0 or r >=N or c <0 or c>= M:
        return

    #    
    if visited[r][c] or S[r][c] == '1':
        return

    #       0.
    visited[r][c] = True

    # , , ,  
    explore(r - 1, c, S, visited, N, M) # 
    explore(r + 1, c, S, visited, N, M) # 
    explore(r, c - 1, S, visited, N, M) # 
    explore(r, c + 1, S, visited, N, M) # 

S = [
 "00110",
 "00011",
 "11111",
 "00000"
]


solution(S, 4, 5)
```
