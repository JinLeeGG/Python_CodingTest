# 동전 문제

```python
n = 1260
cnt = 0
# 가장 큰 화폐 순서대로
coins = [500, 100, 40, 10]

# 화폐 단위가 큰 순서대로 반복
for coin in coins:

    # 현재 화폐 단위로 거슬러 줄 수 있는 개수 계산
    cnt += n // coin

    # 거스르고 남은 금액 갱신
    n %= coin

# 마지막에 처리한 화폐 단위 출력 (이 부분은 의도에 따라 변경 가능)
print(coin)
```

```python
def solution(number, k):
    stack = []
    for digit in number:
        # While stack is not empty, k > 0, and the last element in stack is less than the current digit
        while stack and k > 0 and stack[-1] < digit:
            stack.pop() # Remove the last element from the stack
            k -= 1 # Decrement k as we removed a digit
        stack.append(digit) # Add the current digit to the stack

    # If k is still greater than 0, remove the remaining k digits from the end of the stack
    while k > 0:
        stack.pop()
        k -= 1

    return "".join(stack) # Join the elements in the stack to form the largest number

print(solution("1924", 2)) # Example usage with k=2 to remove two digits
print(solution("1231234", 3)) # Example usage with k=3
print(solution("4177252841", 4)) # Example usage with k=4
```
