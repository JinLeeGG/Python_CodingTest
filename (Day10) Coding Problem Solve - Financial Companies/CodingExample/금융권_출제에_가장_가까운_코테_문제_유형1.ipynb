```python
def solution(n, k):
    # Convert n to base k
    a = ''
    while n > 0:
        d = n % k
        n = n // k
        a += str(d)
    a = a[::-1]  # Reverse to get correct base k representation

    # Sum non-empty segments after splitting by '0'
    c = 0
    for b in a.split('0'):
        if b != '':
            c += int(b)  # Treat each segment as base-10 number

    # Convert sum back to base k
    ret = ''
    while c > 0:
        d = c % k
        c = c // k
        ret += str(d)
    ret = ret[::-1]  # Reverse to get correct base k representation

    # Convert the base k string to an integer (assuming base 10 output)
    return int(ret, k) if ret else 0

# Read input
n, k = map(int, input().split())
print(solution(n, k))
```
