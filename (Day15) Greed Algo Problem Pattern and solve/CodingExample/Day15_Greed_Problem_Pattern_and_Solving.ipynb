```python
def solution(money):
    # 변수 초기화
    n = len(money)
    dp1 = [0] * n           # dp1, dp2의 크기 n만큼 생성
    dp2 = [0] * n

    # 첫번째 집을 도둑질 했을 때
    dp1[0] = money[0]
    dp1[1] = money[0]

    # 첫번째 집을 도둑질하면 마지막집은 도둑질하지 못한다.
    for i in range(2, n-1):
        dp1[i] = max(dp1[i-1], dp1[i-2] + money[i])

    # 첫번째 집을 도둑질하지 않았을때의 구현
    dp2[1] = money[1]
    for i in range(2, n-1):
        dp2[i] = max(dp2[i-1], dp2[i-2] + money[i])

    # 두 경우 중 최대값 구하기
    answer = max(dp1[-2], dp2[-1])
    return answer

solution([1, 2, 3, 1])
```

```python
# My Answer
people = [70, 50, 80, 50, 30]
limit = 100

def solution(people, limit):
    answer = 0
    count = 0
    people.sort()
    for i in range(len(people)):
        for j in range(i+1, len(people)):
            if people[i] + people[j] <= limit:
                count += 1

    answer += len(people)

    return answer

solution(people, limit)
```

```python
def solution(people, limit):
    people.sort()
    cnt = 0
    i = 0 # index
    k = len(people) - 1 # index of last person

    while i <= k:
        if people[k] + people[i] <= limit:
              i += 1
        k -= 1
        cnt += 1
    return cnt

people = [70, 50, 80, 50, 30]
limit = 100

solution(people, limit)
```

```python
def solution(cap, n, deliveries, pickups):
    ans = 0

    # 먼곳부터 순차적으로 배달
    deliveries = deliveries[::-1]
    pickups = pickups[::-1]

    delivery = 0
    pickup = 0
    for i in range(n):
        delivery += deliveries[i]
        pickup += pickups[i]

        while delivery > 0 or pickup > 0:
              delivery -= cap
              pickup -= cap
              ans += (n - i)

    return ans * 2 # 왕복

solution(4,	5,	[1, 0, 3, 1, 2],	[0, 3, 0, 4, 0])
```
