```python
# My Answer
def solution(N, A, B):
    a = []
    index = []
    for i in range(1, N+1, 2):
      a.append((i, i+1)) # make an arry of competiters and group them into tuple

    for i in range(len(a)):
      if A in a[i] and B in a[i]:  # if current tuple contains both A, B
        return 1  # it means they meet in the round 1
      elif A in a[i] or B in a[i]: # elif go through tuple and find index of tuple
        index.append(i) # append it to index list

    index.sort(reverse=True) # sort them decreasing order

    return index[0] - index[1] + 1 # calculate the round

solution(8, 4, 7)
```

```python
def next_num(num):
    return num // 2 if num % 2 == 0 else (num + 1) // 2

def check_meet(a,b):
    return sorted([a, b])[0] % 2 == 1 and abs(a - b) == 1

def solution(n, a, b):
    answer = 1
    while True:
        result = check_meet(a, b)
        if result == True:
          break
        else:
          a = next_num(a)
          b = next_num(b)
          answer += 1

    return answer

solution(8, 4, 7)
```
