```python
# n as a change
# m as a list of money

def change(n, money):
  index = 0
  way = 0
  if money[index] > n:
    way += 0
  elif n % money[index]:
    way += 1
  else:
    return way

change(5, [1,2,5])

```

```python
def fib(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)


```

```python
# d is a list of different department's budget
# b is a budget
# Go through list of different department starting in index 0 and find every combination
# that adds to budget b


def function(d, b):
  pointer1 = 1
  combination = 0
  i = 0
  for i in d:
    if d[i] + d[pointer1] <= b:
      combination +=1
      if i == len(d):
        i +=1
    return combination

function([1,2,3,4,5], 9)


```
